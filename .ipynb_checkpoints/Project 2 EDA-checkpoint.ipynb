# EDA: Check for outliers

We need to check for outliers for the Sales Price of the houses. First, we find the general statistics of the salesprice.

```python
df_train['saleprice'].describe().round(2)
```

We have the general statistics of the salesprice now. A box plot visualisation will help us spot any outliers.

Let us also note that the mean price for houses is $181469.70, which means our baseline model will be based on this figure as the baseline model takes into account outliers as well.

```python
plt.figure(figsize = (15,8))
sns.set(font_scale = 2)
sns.boxplot(x = df_train['saleprice']).set_title('Box Plot of Sales Price of AMES Houses');
```

Seems there are quite a number of outliers. I will filter out the columns with salesprice that differs from the mean sales price by more than 3 standard deviation increments.

```python
sd = np.std(df_train['saleprice']) #standard dev of train data
mean = np.mean(df_train['saleprice']) #mean of train data
```

```python
df_train.drop(df_train[df_train.saleprice > (mean + 3*sd)].index, inplace = True) #remove data of salesprice > mean + 3*sd
df_train.drop(df_train[df_train.saleprice < (mean - 3*sd)].index, inplace = True) #remove data of salesprice < mean - 3*sd
```

```python
plt.figure(figsize = (15,8))
sns.set(font_scale = 2)
sns.boxplot(x = df_train['saleprice']).set_title('Box Plot of Sales Price of AMES Houses (Outliers Removed)');
```

We have managed to remove the outliers on the salesprice over 3 standard deviation from the mean saleprice. For now we will continue cleaning our data and monitor for any other potential issues.

Let's save the outlier's removed DataFrame as another csv so that the modeling and prediction notebook is able to continue reading data.

```python
df_train.to_csv('datasets/noutlier_clean_train.csv')
```

### Feature Engineer our Garage  
Garage seem to have a rather moderate correlation to saleprice. Also from the sneak EDA during the data cleaning process, I noticed that several Garage columns are strongly correlated to each other.   
Instead of removing those strongly correlated data,  perhaps we can create a feature of garage to see if it helps. Let's feature our garage finish, garage quality, garage condition, garage area, garage year built and garage cars as an overall garage score feature

```python
df_garage_overall = pd.DataFrame()
df_garage_overall['garage_overall'] = df_train['garage_finish']*df_train['garage_qual']*df_train['garage_cond']*df_train['garage_area']*df_train['garage_cars']*df_train['garage_yr_blt']
# Get our features

df_garage_overall['saleprice'] = df_train['saleprice']
# Set another column to saleprice

sns.set(rc = {'figure.figsize':(15,8)}, font_scale = 1.5)
sns.heatmap(df_garage_overall.corr(),annot = True).set_title('Heatmap of Garage Overall Score Feature');
# Get our correlation number and heatmap
```

It appears we have a higher correlation by featuring garage overall score as a product of most of the garage features. Let's investigate further.

```python
sns.boxplot(x = df_garage_overall['garage_overall']).set_title('Garage Overall Score');
```

```python
sns.set(rc = {'figure.figsize':(15,15)}, font_scale = 1)

sns.pairplot(df_garage_overall);
```

```python
df_garage_overall['garage_overall'].describe() # Get statistics of our garage overall score
```

Wow it appears there is also outliers in this data despite us removing outliers of saleprice. However, we will not remove these outliers as we are not modeling after garage score, we are modeling for sale price of the house. However this does show the general trend of our garage, that it does have a relatively large impact on our saleprice. This is to be expected as AMES is a location beside the highway, and is also known to be an agricultural area, thus the need for a garage.
